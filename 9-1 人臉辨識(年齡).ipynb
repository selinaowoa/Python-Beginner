```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import cv2
```

```python
face_cascade = cv2.CascadeClassifier('d:/ipython/cv2/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('d:/ipython/cv2/haarcascade_eye.xml')

img = cv2.imread('d:/ipython/chen3.jpg')
```

```python
gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
import random
font=cv2.FONT_HERSHEY_SIMPLEX
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    cv2.putText(img, str(random.randrange(40,45)), (x+(w//2)-18, y-10), font, 1, (14, 201, 255), 2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)

cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
